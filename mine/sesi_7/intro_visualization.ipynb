```python
import numpy as np
import pandas as pd
```

```python
df_canada=pd.read_excel("Canada.xlsx",sheet_name="Canada by Citizenship",skiprows=20)
df_canada.head()
```

### Visualizing the data

```python
import matplotlib.pyplot as plt
```

```python
df_canada.head()
```

```python
df_canada.rename(columns={'OdName':'Country','AreaName':'Continent'},inplace=True)
```

```python
df_canada.head()
```

```python
df_canada.set_index('Country',inplace=True)
```

```python
df_canada
```

```python
tahun=list(range(1980,2014))
s_canada_haiti=df_canada.loc['Haiti',tahun]
s_canada_haiti.head()
```

```python
s_canada_haiti.plot();
```

```python
s_canada_haiti.plot(color='red');

plt.title("Imigrasi dari Haiti");
plt.xlabel("Tahun");
plt.ylabel("Jumlah Imigran");
plt.text(2003,6000,"Gempa 2010");
```

```python
china_india=df_canada.loc[['China','India'],tahun]
china_india.head()
```

```python
china_india.plot();
```

```python
china_india.T.plot(color=['red','orange']);

plt.title("Imigrasi dari China dan India");
plt.xlabel("Tahun");
plt.ylabel("Jumlah Imigran");
```

5 Negara Terbesar Penyumbang Imigran Ke Jerman tahun 1980-2013

```python
s_total=df_canada[tahun].sum(axis=1)
df_canada['Total']=s_total
df_canada.sort_values(['Total'],ascending=False,inplace=True)
```

```python
df_canada.drop(['Total','Unknown'],inplace=True)
```

```python
df_canada[tahun].head().T.plot(kind="area",alpha=0.5,stacked=False,figsize=(15,8));

plt.title("Top 5 Negara Penyumbang imigran Terbanyak");
plt.xlabel("Tahun");
plt.ylabel("Jumlah Imigran");
```
