```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
import numpy as np
import pandas as pd
dadu=np.random.randint(1,7,size=100)
pd.Series(dadu).value_counts().plot(kind='bar');

plt.yticks()
```

```python
from scipy.stats import uniform

n=10000
start=10
width=20
data_uniform=uniform.rvs(loc=start,scale=width,size=n)
```

```python
pd.Series(data_uniform).plot(kind='hist',bins=100);
```

```python
ax=sns.distplot(
    data_uniform,
    bins=100,
    color='skyblue',
    hist_kws={'alpha':1,'linewidth':15},
)

ax.set(xlabel='Data',ylabel='Frequency');
```
