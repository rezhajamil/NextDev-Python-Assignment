### Data Kejahatan di San Fransisco

```python
df_incidents=pd.read_csv("https://github.com/ardhiraka/PFDS_sources/raw/master/Police_Department_Incidents_-_Previous_Year__2016_.csv")
df_incidents.head()
```

```python
df_incidents.shape
```

```python
df_incidents=df_incidents.head(100)
df_incidents.shape
```

```python
longitude=-122.42
latitude=37.77

sf_map=folium.Map(location=[latitude, longitude], zoom_start=15)
sf_map
```

```python
from folium import plugins

sf_map=folium.Map(location=[latitude, longitude], zoom_start=15)
incidents=plugins.MarkerCluster().add_to(sf_map)

latitudes=list(df_incidents.Y)
longitudes=list(df_incidents.X)
labels=list(df_incidents.Category)

for lat,long,label in zip(df_incidents.Y, df_incidents.X,df_incidents.Category):
    folium.Marker(
        location=[lat, long], 
        popup=label, 
        icon=folium.Icon(color='orange')
    ).add_to(incidents)

sf_map
```
